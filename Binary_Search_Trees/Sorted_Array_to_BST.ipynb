```python
from PredefinedBSTs import *

class BSTNode:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

def print_binary_tree(root):
    if(root is None):
        return

    print(root.data,end = ": ")

    if(root.left is not None):
        print(f"L->{root.left.data}",end = ", ")
    else:
        print("L->None",end = ",")

    if(root.right is not None):
        print(f"R->{root.right.data}")
    else:
        print("R->None") 

    print_binary_tree(root.left)
    print_binary_tree(root.right)


def sortedListToBST(l1):
    if(len(l1) == 0):
        return None

    mid = len(l1)//2 # Finding the middle index

    rootData = l1[mid]
    root = BSTNode(rootData)

    root.left = sortedListToBST(l1[:mid])
    root.right = sortedListToBST(l1[mid + 1:])
    return root

root = sortedListToBST([1,2,3,4,5,6,7])
print_binary_tree(root)
```
