```python
from Common import CreateLLFromList,print_LL,Node

head = CreateLLFromList([1,2,3,4,5])

print_LL(head)

def search_by_value(head,value):
    temp = head
    index = 0

    while temp != None:
        if(temp.data == value):
            return index
        temp = temp.next
        index += 1

    return "Not Found"

print("Searching")
print(search_by_value(head,5))
```

```python
def search_by_index(head, index):
    if index < 0:
        return None
    
    current = head
    count = 0
    
    while current is not None:
        if count == index:
            return current.data
        current = current.next
        count += 1
    
    return None
```
